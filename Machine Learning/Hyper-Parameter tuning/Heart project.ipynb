```python
import pandas as pd
```

```python
df = pd.read_csv("heart.csv")
df.head(3)
```

```python
df.info()
```

```python
categorical_data = df.select_dtypes(include = ["object"])
numerical_data = df.select_dtypes(include = ["number"])

categorical_data, numerical_data
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[["Age", "RestingBP", "Cholesterol", "FastingBS", "MaxHR", "Oldpeak"]] = scaler.fit_transform(df[["Age", "RestingBP", "Cholesterol", "FastingBS", "MaxHR", "Oldpeak"]])
df.head(3)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
column_transformer = ColumnTransformer(
    transformers = [
        ("Sex", OneHotEncoder(), ["Sex"]),
        ("ChestPainType", OneHotEncoder(), ["ChestPainType"]),
        ("RestingECG", OneHotEncoder(), ["RestingECG"]),
        ("ExcerciseAngina", OneHotEncoder(), ["ExerciseAngina"]),
        ("ST_Slope", OneHotEncoder(), ["ST_Slope"])
    ],
    remainder = "passthrough"
)
df = column_transformer.fit_transform(df)
df = pd.DataFrame(df, columns = column_transformer.get_feature_names_out())
df.head(3)
```

```python
X = df.drop("remainder__HeartDisease", axis = 1)
y = df["remainder__HeartDisease"]
X.shape, y.shape
```

```python
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import LogisticRegression, Lasso, Ridge

model_params = {
    "DecisionTreeClassifier":{
        "model":DecisionTreeClassifier(),
        "params":{
            "criterion":["gini", "entropy"]
        }
    },
    "RandomForestClassifier":{
        "model":RandomForestClassifier(),
        "params":{
            "criterion":["gini", "entropy"],
            "oob_score":[True, False],
            "n_estimators":[i for i in range(20, 100, 25)]
        }
    },
    "SVC":{
        "model":SVC(),
        "params":{}
    },
    "LogisticRegression":{
        "model":LogisticRegression(max_iter = 100000),
        "params":{
            "tol":[i/100 for i in range(1, 1000, 50)],
            "C":[i/10 for i in range(0,100,25)]
        }
    },
    "LassoRegularization":{
        "model":Lasso(max_iter = 100000),
        "params":{
            "tol":[i/100 for i in range(0, 1000, 50)],
            "alpha":[i/10 for i in range(1, 11)]
        }
    }
}
score = []
for model_name, mp in model_params.items():
    clf = RandomizedSearchCV(mp["model"], mp["params"], cv = 5, return_train_score = False, n_iter = 10000)
    clf.fit(X, y)
    score.append({"model_name":[model_name], "best_scores":[clf.best_score_], "best_parameters":[clf.best_params_]})

df = pd.DataFrame(score)
df.head(5)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 20)
y1 = np.sin(x)
y2 = np.sin(x - np.pi)
```

```python
with plt.style.context("fivethirtyeight"):
    plt.figure()
    plt.plot(x, y1)
    plt.plot(x, y2)
```

```python
with plt.style.context("fivethirtyeight"):
    plt.figure()
    plt.plot(x,
             y1,
             color = "black",
             linestyle ="-",
             linewidth = 2,
             marker = "s",
             markersize = 6,
             label = "y1")
    plt.plot(x,
             y2,
             color = "gray",
             linestyle ="--",
             linewidth = 2,
             marker = "^",
             markersize = 6,
             label = "y2")
    plt.legend()
```

```python
years = np.arange(2012, 2015)
c1_values = [2, 5, 9]
c2_values = [7, 6, 3]
```

```python
plt.figure()
plt.bar(years, c1_values)
```

```python
with plt.style.context("fivethirtyeight"):
    #note - we use the value 0.2 since it is half the width of the bars
    plt.bar(years + 0.2
            , c1_values
            , color = "blue"
            , edgecolor = "red",
            width = 0.4,
            align = "center",
            label = "y1")
    plt.bar(years - 0.2
            , c2_values
            , color = "orange"
            , edgecolor = "red",
            width = 0.4,
            align = "center",
            label = "y1")
    plt.legend()
    plt.xticks(years, [str(year) for year in years])
```

```python
categories = ["A", "B", "C", "D", "E"]
values = [7, 12, 4, 2, 9]
```

```python
with plt.style.context("fivethirtyeight"):
    plt.figure()
    plt.barh(np.arange(len(categories)), values)
    plt.yticks(np.arange(len(categories)), ["category{}".format(x) for x in categories])
```

```python
with plt.style.context("fivethirtyeight"):
    plt.figure()
    plt.barh(-np.arange(len(categories)), values)
    plt.yticks(-np.arange(len(categories)), ["category{}".format(x) for x in categories])
```

```python
with plt.style.context("fivethirtyeight"):
    plt.figure()
    plt.barh(np.arange(len(categories)), values,
             color = "blue",
             edgecolor = "red",
             height = 0.4,
             align = "center",
             label = "y1")
    plt.yticks(np.arange(len(categories)), ["category{}".format(x) for x in categories])
```

```python
counts = [17, 14]
```

```python
with plt.style.context("fivethirtyeight"):
    plt.figure(figsize = (4, 4))
    plt.pie(counts)
```

```python
with plt.style.context("fivethirtyeight"):
    plt.figure(figsize = (4, 4))
    plt.pie(counts,
           colors = ["blue", "orange"],
           labels = ["category A", "Category B"],
           startangle = 90,
           autopct = "%1.1f%%")
```

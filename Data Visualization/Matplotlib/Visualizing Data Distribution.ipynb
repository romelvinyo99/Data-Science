```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
column_data = np.random.normal(42, 3, 1000)
```

```python
with plt.style.context("fivethirtyeight"):
    plt.figure()
    plt.hist(column_data)
```

```python
with plt.style.context("fivethirtyeight"):
    plt.figure()
    plt.hist(column_data,
            color = "red",
            bins = 50)
```

```python
two_column_of_data = [
    (np.random.normal(57, 3, 1000)),
    (np.random.normal(42, 7, 1000))
]
```

```python
with plt.style.context("fivethirtyeight"):
    plt.figure()
    plt.boxplot(two_column_of_data)
    
```

```python
with plt.style.context("fivethirtyeight"):
    plt.figure()
    plt.boxplot(two_column_of_data,
               labels = ["A", "B"],
               notch = True,
               bootstrap = 1000)
```

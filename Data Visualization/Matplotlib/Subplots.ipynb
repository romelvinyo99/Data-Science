```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
with plt.style.context("fivethirtyeight"):
    plt.figure()
    for i in range(1, 7):
        plt.subplot(3, 2, i)
        plt.title(i)
        plt.xticks([])
        plt.yticks([])
plt.tight_layout()        
```

```python
dist1 = np.random.normal(42, 7, 1000)
dist2 = np.random.normal(59, 3, 1000)

plt.figure(figsize = (10, 4))
#Row 1 column 1
plt.subplot(1, 2, 1)
plt.hist(dist1)
plt.title("dist1")
#Row 1 column 2
plt.subplot(1, 2, 2)
plt.scatter(dist2, dist1)
plt.xlabel("dist2")
plt.ylabel("dist1")
#Makes sure the plots do not overlap
plt.tight_layout
```

```python
years = np.arange(2010, 2016)

plt.figure()
#we are pltting lines charts for 16 different values in the same figure which is inneficient
for category_num in range(1, 17):
    y_vals = np.arange(10, 16) + (np.random.random(6)*category_num / 4)
    plt.plot(years, y_vals)
plt.ylim(10, 18)
plt.xticks(years, [str(years) for year in years])
```

```python
years = np.arange(2010, 2016)

plt.figure(figsize = (20,20))

for category_num in range(1, 17):
    plt.subplot(4, 4, category_num)
    y_vals = np.arange(10, 16) + (np.random.random(6) * category_num / 4)
    plt.plot(years, y_vals)
    plt.ylim(10, 18)
    plt.xticks(years, [str(years) for year in years])
    plt.title("Category {}".format(category_num))
```

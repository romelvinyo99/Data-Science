```python
#importing the necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
#importing the dataset
df = pd.read_csv("HR_comma_sep.csv")
df
```

```python
#Cleaning the dataset
x = df.info()
y = df.isna().shape
z = df.duplicated()
x, y, z
```

```python
categorical_data = df.select_dtypes(include = ["object"])
numerical_data = df.select_dtypes(include = ["number"])
categorical_data, numerical_data
```

```python
#...summary statistics... - numerical_data
a = df.describe(include = ["number"], exclude = ["object"])
b = numerical_data.median()
c = numerical_data.mode()
a, b, c
```

```python
#...One hot encoding...
le = LabelEncoder()
categorical_data_transformed = categorical_data.apply(le.fit_transform)
categorical_data_transformed
```

```python
df_updated = df.drop(columns = categorical_data.columns)
df_transformed = pd.concat([df_updated, categorical_data_transformed], axis = "columns")
df_transformed
```

```python
#...summary statistics - categorical columns
a = categorical_data_transformed.describe()
b = categorical_data_transformed.median()
c = categorical_data_transformed.mode()
a, b, c
```

```python
#creating regression lines for each column against leaving
X = df_transformed.drop(columns = ["left"])
y = df_transformed["left"]
correlation_matrix = df_transformed.corr()["left"].drop(["left"]).sort_values(ascending=False)
cm_mean = correlation_matrix.mean()
cm_median = correlation_matrix.median()
correlation_matrix, cm_mean, cm_median
```

```python
# Employee retention - based on salary

print(df_transformed["salary"].unique())

salary_retention = df_transformed.groupby("salary")["left"].count()
fig, ax = plt.subplots(figsize = (12, 6))
salary_retention.plot(kind = "bar", alpha = 0.5, ax = ax)
for rect in ax.patches:
        height = rect.get_height()
        ax.annotate(
            f"{height:.0f}",
            xy = (rect.get_x()+rect.get_width()/2, height),
            xytext = (0, 3),
            textcoords = "offset points",
            ha = "center",
            va = "bottom"
        )
plt.xlabel("salary")        
plt.ylabel("left")        
plt.legend()        
plt.tight_layout()
plt.show()

print(df_transformed["salary"].unique())
df.salary.unique()
```

```python
# Employee retention - based on department

print(df_transformed["Department"].unique())

salary_retention = df_transformed.groupby("Department")["left"].count()
fig, ax = plt.subplots(figsize = (12, 6))
salary_retention.plot(kind = "bar", alpha = 0.5, ax = ax)
for rect in ax.patches:
        height = rect.get_height()
        ax.annotate(
            f"{height:.0f}",
            xy = (rect.get_x()+rect.get_width()/2, height),
            xytext = (0, 3),
            textcoords = "offset points",
            ha = "center",
            va = "bottom"
        )
plt.xlabel("department")        
plt.ylabel("left")        
plt.legend()        
plt.tight_layout()
plt.show()

print(df_transformed["Department"].unique())
df.Department.unique()
```

```python
#importing the required libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing the dataset - dataframe(df)
df = pd.read_csv("supermarket_sales - Sheet1.csv")

#printing the first five elements of the dataset
print(df.head(5))

#printing the last five elements of the dataset
print(df.tail(5))

```

```python
#cleaning the datatypes - implicit errors

#checking for the summary information of the dataset below
print(df.info())
"""
From the information below we can deduce:
1.shape  = 1000rows by 17columns (1000, 17)
2.There are no null values in all the entries
3.We can see the datatypes of the values:
     1.floats  = 7
     2.Strings = 9 
     3.Integer = 1
"""
#checking for duplicate values
duplicated_rows = df[df.duplicated()]
print(duplicated_rows)
"""
From the above information we can see that there are no duplicate rows
This is because we returned an empty array with no values
otherwise
df.drop_duplicates()
"""
```

```python
#Data cleaning - explicit
"""
revenue = quantity * price
cogs = revenue * tax_rate
gross income = revenue - cogs
gross margin percentage = ((revenue - cogs)/revenue) * 100
Total = revenue + tax
"""

df["cogs"] = df["Quantity"] * df["Unit price"] * 0.05
df["Revenue"] = df["Quantity"] * df["Unit price"]
df["gross income"] = df["Revenue"] - df["cogs"]
df["gross margin percentage"] = ((df["Revenue"] - df["cogs"])/df["Revenue"]) * 100
```

```python
#summary statistics
df.describe()

#getting the median for the iterable variables
df[["Unit price", "Total", "Quantity", "Tax 5%", "cogs", "gross margin percentage", "gross income", "Rating"]].median()

"""
From the information below we can get the:median
                                          count
                                          mean
                                          standard deviation
                                          minimum
                                          maximum
                                          quartiles
"""
```

```python
#Question 1 - Which product line is most popular and with which gender

product_gender = df.groupby(["Gender","Product line"])["Quantity"].count()


fig, ax = plt.subplots(figsize=(6,6))
bars = product_gender.unstack().plot(kind="bar", ax=ax)
for bar in bars.containers:
    for rect in bar.patches:
        height = rect.get_height()
        ax.annotate(
            f"{height:.0f}",
            xy = (rect.get_x()+rect.get_width()/2, height),
            xytext = (0,3),
            textcoords = "offset points",
            ha = "center",
            va = "bottom"
        )
plt.title("product line distribution by gender")
plt.xlabel("Product line")
plt.ylabel("product line")
plt.tight_layout()
plt.show()
```

```python
#Which product line is most popular

product_gender = df.groupby("Product line")["Quantity"].count()

fig, ax = plt.subplots(figsize=(6,6))
product_gender.plot(kind = "bar", color = ["Yellow", "pink"], alpha=0.5)
for rect in ax.patches:
    height = rect.get_height()
    ax.annotate(
        f'{height:.0f}', 
        xy = (rect.get_x()+rect.get_width()/2, height), 
        xytext = (0, 3),
        textcoords = "offset points", 
        ha = "center", 
        va = "bottom"
    )
plt.title("product popularity")
plt.xlabel("gender/ product")
plt.ylabel("count")
plt.show()

```

```python
#Question 2 - which gender has the highest total rating and the mean and median ratings

gender_rating_comparison = df.groupby("Gender")["Rating"].agg(["mean", "median", "count"])
fig, ax = plt.subplots(figsize = (6, 6))
gender_rating_comparison.plot(kind = "bar",subplots = False, ax= ax, color = ["green", "red", "yellow"], alpha=0.5)
for rect in ax.patches:
    height = rect.get_height()
    ax.annotate(
        f'{height:.0f}', 
        xy = (rect.get_x() + rect.get_width()/2, height), 
        xytext = (0, 3) ,
        textcoords = "offset points", 
        ha = "center", 
        va = "bottom"
    )
plt.title("mean and median Gender agaist rating")
plt.xlabel("gender")
plt.ylabel("count/mean/median")
plt.show()
```

```python
# Question 3 - which branch has the highest, mean and median rating depending on the city and the membership
branch_city_rating = df.groupby(["Branch", "City", "Customer type"])["Rating"].agg(["mean", "median", "count"])
fig, ax = plt.subplots(figsize = (12, 16))
bars = branch_city_rating.unstack().plot(kind = "bar", ax = ax, alpha = 0.7)
plt.xlabel("branch")
plt.ylabel("rating")
for bar in bars.containers:
    for rect in bar.patches:
        height = rect.get_height()
        ax.annotate(
               f'{height:.0f}',  
            # Position of the annotation (center of the bar)
            xy=(rect.get_x() + rect.get_width() / 2, height),
            # Position of the text label relative to the annotation (3 points above)
            xytext=(0, 3),  
            # Offset points relative to text coordinates
            textcoords="offset points",
            # Horizontal alignment of the text
            ha='center',
            # Vertical alignment of the text
            va='bottom'
        )
```

```python
#Question 4 - which city brings in the highest income for the company and the mean and median of the income generated by the genders

city_gross = df.groupby("City")["gross income"].agg(["mean", "median", "count"])
fig, ax = plt.subplots(figsize = (12, 12))
bar = city_gross.unstack().plot(kind = "bar", ax = ax, color=["red", "orange", "yellow", "pink", "blue"], alpha=0.5)
plt.title("city mean and median gross income")
plt.ylabel("gross income")
plt.xlabel("city")
for rect in ax.patches:
    height = rect.get_height()
    ax.annotate(
        f'{height:.0f}',
        xy = (rect.get_x()+rect.get_width()/2, height),
        xytext = (0,3),
        textcoords = "offset points",
        ha = "center",
        va = "bottom"
    )
plt.tight_layout()    
plt.show()
```

```python
#which city brings in the highest gross income

city_gross = df.groupby("City")["gross income"].sum()

fig, ax = plt.subplots(figsize = (12, 12))
bar = city_gross.plot(kind = "bar", ax = ax, color=["red", "orange", "yellow", "pink", "blue"], alpha=0.5)
plt.title("city gross income return")
plt.ylabel("gross income")
plt.xlabel("city")
for rect in ax.patches:
    height = rect.get_height()
    ax.annotate(
        f'{height:.0f}',
        xy = (rect.get_x()+rect.get_width()/2, height),
        xytext = (0,3),
        textcoords = "offset points",
        ha = "center",
        va = "bottom"
    )
plt.tight_layout()    
plt.show()
```

```python
#Question 5 - The customer status of people in each city and what is their gender
city_member_gender = df.groupby(["Gender", "Customer type", "City"]).size()
fig, ax = plt.subplots(figsize = (12, 6))
bar = city_member_gender.unstack().plot(kind = "bar", ax=ax)
for bars in bar.containers:
    for rect in bars.patches:
        height = rect.get_height()
        ax.annotate(
            f"{height:.0f}",
            xy = (rect.get_x()+rect.get_width()/2, height),
            xytext = (0,3),
            textcoords = "offset points",
            ha = "center",
            va = "bottom"
        )
plt.yscale("log")
plt.tight_layout()
plt.show()
```

```python
#Question 6 - what is the consumption quantity of products in each city


city_productline = df.groupby(["City", "Product line"])["Quantity"].count()
fig, ax = plt.subplots(figsize = (15, 13))
bars = city_productline.unstack().plot(kind = "bar", ax = ax)
for bar in bars.containers:
    for rect in bar.patches:
        height = rect.get_height()
        ax.annotate(
               f'{height:.0f}',  
            # Position of the annotation (center of the bar)
            xy=(rect.get_x() + rect.get_width() / 2, height),
            # Position of the text label relative to the annotation (3 points above)
            xytext=(0, 3),  
            # Offset points relative to text coordinates
            textcoords="offset points",
            # Horizontal alignment of the text
            ha='center',
            # Vertical alignment of the text
            va='bottom'
        )
        
```

```python
#Question 7 - Which gender are the highest consumers in each city which products do they buy
gender_city_products = df.groupby(["Gender", "City", "Product line"])["Quantity"].count()
fig, ax = plt.subplots(figsize=(12,6))
bar = gender_city_products.unstack().plot(kind="bar", ax=ax)
plt.ylabel("Quantity")
for bars in bar.containers:
    for rect in bars.patches:
        height=rect.get_height()
        ax.annotate(
            f"{height:.0f}",
            xy = (rect.get_x()+rect.get_width()/2, height),
            xytext = (0,3),
            textcoords = "offset points",
            ha = "center",
            va = "bottom"
        )
plt.tight_layout()        
plt.show()        
        
```

```python
#Question 8 - Which city and gender pay the highest taxes

gender_city_taxes = df.groupby(["Gender", "City"])["Tax 5%"].agg(["mean", "median", "count"])
fig, ax = plt.subplots(figsize=(14,14))
bar = gender_city_taxes.unstack().plot(kind = "bar", alpha=0.6, ax=ax)
plt.ylabel("tax")
for bars in bar.containers:
    for rect in bars.patches:
        height=rect.get_height()
        ax.annotate(
            f"{height:.0f}",
            xy = (rect.get_x()+rect.get_width()/2, height),
            xytext = (0,3),
            textcoords = "offset points",
            ha = "center",
            va = "bottom"
        )
plt.tight_layout()        
plt.show()        
```

```python
#Question 9 - what is the most preferred type of payment in each city and by which gender

payment_city_gender = df.groupby(["City", "Gender", "Payment"]).size()
fig, ax = plt.subplots(figsize = (30,30))
bar = payment_city_gender.unstack(level=0).plot(kind = "pie",subplots=True,autopct='%1.1f%%', ax=ax)
plt.ylabel("count")       
plt.legend(prop={"size": 8}, loc="lower right", prop={"size":8},loc = "lower right",prop={"size":1}, loc = "lower right")
plt.show()
        
```

```python
# Question 10 - which gender generates more income for the company

grossincome_gender = df.groupby("Gender")["gross income"].agg(["count", "median", "mean"])
fig, ax = plt.subplots(figsize = (5,5))
grossincome_gender.plot(kind="bar", ax = ax, color = ["blue", "pink"], alpha = 0.5)
for rect in ax.patches:
    height = rect.get_height()
    ax.annotate(
        f'{height:.0f}',
        xy = (rect.get_x()+rect.get_width()/2, height),
        xytext = (0, 3),
        textcoords = "offset points", 
        ha = "center", 
        va = "bottom"
    )
plt.ylabel("gross income")   
plt.tight_layout()
plt.show    
```
